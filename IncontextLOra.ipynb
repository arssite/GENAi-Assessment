```python
import torch
from diffusers import FluxPipeline

pipe = FluxPipeline.from_pretrained("black-forest-labs/FLUX.1-schnell", torch_dtype=torch.bfloat16)
pipe.enable_model_cpu_offload() #save some VRAM by offloading the model to CPU. Remove this if you have enough GPU power

prompt = "A cat holding a sign that says hello world"
image = pipe(
    prompt,
    guidance_scale=0.0,
    num_inference_steps=4,
    max_sequence_length=256,
    generator=torch.Generator("cpu").manual_seed(0)
).images[0]
image.save("flux-schnell.png")
```

```python
import torch
from diffusers import FluxPipeline

# This line will check if CUDA is available and use it if possible.
# If not, it will default to CPU.
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

pipe = FluxPipeline.from_pretrained("black-forest-labs/FLUX.1-schnell", torch_dtype=torch.bfloat16)
# Explicitly move the pipeline to the determined device
pipe = pipe.to(device)

# If you have enough VRAM, remove 'pipe.enable_model_cpu_offload()'
pipe.enable_model_cpu_offload() # Save VRAM by offloading the model to CPU.

prompt = "A cat holding a sign that says hello world"
image = pipe(
    prompt,
    guidance_scale=0.0,
    num_inference_steps=4,
    max_sequence_length=256,
    # The generator is explicitly set to the determined device.
    generator=torch.Generator(device).manual_seed(0)
).images[0]
image.save("flux-schnell.png")
```
