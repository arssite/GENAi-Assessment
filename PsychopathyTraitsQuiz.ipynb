```python
def take_psychopathy_quiz():
    questions = [
        "Hurting someone is no big deal and I feel little to no guilt when I feel someone is hurt.",
        "I am very charming and I can always get my way by manipulating people.",
        "I do not feel fear or anxiety in high-stress situations.",
        "I do not hesitate in lying frequently, even when it’s unnecessary.",
        "I prefer power and control over cooperation.",
        "I act on impulse without worrying about the consequences.",
        "I feel detached from people’s emotions/not able to understand them well.",
        "I enjoy taking risks, even when I know they can backfire or land me in legal trouble.",
        "Rules are not to be followed and aren't important.",
        "I get bored easily and my brain needs constant stimulation.",
        "I am not afraid to hurt others if it helps me achieve my goals.",
        "I lack empathy for most people.",
        "I think emotions make people weak and vulnerable.",
        "I get a sense of satisfaction when I watch others struggle.",
        "I feel I am superior to most people.",
        "I don’t feel bad about manipulating people for my own gain.",
        "I am good at pretending to care about others.",
        "I get angry and argumentative when people challenge or oppose me.",
        "I have trouble forging deep emotional connections with others.",
        "I feel no remorse for actions that harm others if it benefits me."
    ]

    score = 0

    print("Psychopathy Traits Quiz\n")

    for question in questions:
        while True:
            answer = input(f"Q: {question} (Agree/Neutral/Disagree): ").strip().lower()
            if answer == "agree":
                score += 2
                break
            elif answer == "neutral":
                score += 1
                break
            elif answer == "disagree":
                break
            else:
                print("Invalid input. Please answer with Agree, Neutral, or Disagree.")

    print("\nYour total score is:", score)

    if score >= 30:
        print("Results Interpretation: Mostly Agree (30-40 points): You display a high number of psychopathic traits.")
        print("However, this does not prove you are clinically a psychopath. A professional evaluation may be advised.")
    elif score >= 15:
        print("Results Interpretation: Mostly Neutral (15-29 points): You may have some psychopathic traits.")
        print("It’s hard to tell whether you are a clinical psychopath, but these traits aren’t dominant.")
    else:
        print("Results Interpretation: Mostly Disagree (0-14 points): You do not display psychopathic tendencies.")
        print("It looks like you are empathetic, emotionally deep, and have regard for others.")

# Start the quiz
take_psychopathy_quiz()
```
